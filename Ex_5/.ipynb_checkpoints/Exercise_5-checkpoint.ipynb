```python
main_directory = "../data/generated/DWD/hourly/precipitation/historical/"

local_prec_ts_file = "precs_ts_modified_2017.csv"

local_prec_ts_path = main_directory + local_prec_ts_file

station_desc_pattern = "RR_Stundenwerte_Beschreibung_Stationen.csv"

station_desc_filepath = main_directory + station_desc_pattern

local_prec_alt_file_path = main_directory + "precs_altitude.csv"
```

```python
import pandas as pd
import datetime

def prec_ts_csv_to_df(filepath):
    
    df = pd.read_csv(filepath, delimiter=";", encoding="utf8", index_col="date", parse_dates = ["date"], na_values = [-999.0, -999])

    return df
df_from_csv = pd.DataFrame() #dataframe derived from the CSV file
```

```python
df_from_csv = prec_ts_csv_to_df(local_prec_ts_path)

df_from_csv
```

```python
df_precs_sum = pd.DataFrame()
df_precs_sum = df_from_csv.sum()

df_precs_sum
```

```python
df_prec_rate_sum = df_precs_sum.to_frame(name = 'Sum')

#df_prec_rate_sum = df_prec_rate_sum.set_index('station_id')

df_prec_rate_sum.index.rename(name = "station_id", inplace = True)

df_prec_rate_sum
```
