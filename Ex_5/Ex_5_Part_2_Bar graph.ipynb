# Exercise 5: Part 2 - Plotting the altitude bar graph

```python
#Path of main directory
main_directory = "../data/generated/Exercise 5/"

#Csv containing altitudes of weather stations and DTM was generated from QGIS
#Csv file containing altitude data
alt_csv = "Station_alt_vs_DTM_altitude.csv"

#File path of CSV file
alt_csv_path = main_directory + alt_csv
```

```python
#Converting csv to Pandas dataframe

import pandas as pd
import datetime

def alt_csv_to_df(filepath):
    
    df = pd.read_csv(filepath, delimiter=",", encoding="utf8", index_col="station_id", usecols = ["station_id", "altitude", "rvalue_1"], na_values = [-999.0, -999])

    return df

#dataframe containing CSV file data
df_from_csv = pd.DataFrame() 
```

```python
alt_df_from_csv = alt_csv_to_df(alt_csv_path)

alt_df_from_csv
```

```python
#Renaming column names of dataframe
alt_df_from_csv.rename(columns={'altitude': 'station altitude', 'rvalue_1': 'DTM altitude'}, inplace=True)
alt_df_from_csv
```

```python
import matplotlib.pyplot as plt
```

```python
#To check the default configurations ny Python
#Python_default = plt.rcParams.copy()

#Python_default
```

```python
#Plotting bar graph for altitude of stations vs DTM altitude

#Changing default figure size

plt.rcParams["figure.figsize"] = (40,30)
#plt.rcParams["legend.fontsize"] = 20
alt_bar = alt_df_from_csv.plot.bar(rot=0)
alt_bar.figure.suptitle('Altitude of stations vs DTM altitude', fontsize=30)
plt.legend(fancybox=True, loc=1, fontsize = 30)
plt.xlabel('Weather stations (Station IDs)', fontsize=28)
plt.ylabel('Altitude (in meters)', fontsize=26)


alt_bar.figure.savefig('Altitude bar chart.jpg')
```
